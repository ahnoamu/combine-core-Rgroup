```python

import numpy as np
import pandas as pd

import rdkit.Chem as Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit.Chem import DataStructs
from rdkit.Chem.MolStandardize import rdMolStandardize
```

# Given

```python

core = "CCN1C(=CC2=CC=CC=[N+]2CC)SC(=C3N(C4=CC=CC=C4S3)C)C1=O"   

Rgroup = "CN"
```

# Method 1
## A. Prepare core

```python

core = Chem.MolFromSmiles("CCN1C(=CC2=CC=CC=[N+]2CC)SC(=C3N(C4=CC=CC=C4S3)C)C1=O", sanitize=False)   
core
```

### &emsp; (i) Label atoms

```python

for atom in core.GetAtoms():
    atom.SetProp("molAtomMapNumber",str(atom.GetIdx()))
core
```

### &emsp;(ii) Identify the target bond for attachment of the Rgroup. Fragment on the bond
##### &emsp;&emsp;&emsp;- R2 should be attached at the bond between atoms 1 and 2 
##### &emsp;&emsp;&emsp;- Fragment on the bond between atoms 1 and 2


```python

bond = core.GetBondBetweenAtoms(1, 2)

print("\nThe bond index of the target bond is ", bond.GetIdx(), "\n")

core_frags = Chem.FragmentOnBonds(core, [bond.GetIdx()], addDummies=True, dummyLabels = [(1, 1)])

core_frags
```

### &emsp;(iii) Obtain the largest fragment (core)
#### &emsp; &emsp;*RDKit automatically inserts a dummy atom on the fragmented bond*

```python

new_core = rdMolStandardize.FragmentParent(core_frags)

new_core
```

### &emsp; (iv) Replace the dummy atom pattern in the core "[ 1 \* ] with "[* : 1]" 

```python
# Current core looks like

print(Chem.MolToSmiles(new_core, kekuleSmiles = True))
```

```python

curr_patt = Chem.MolFromSmiles('[1*]') 
new_patt = Chem.MolFromSmiles('[*:1]')   

new_c = Chem.ReplaceSubstructs(new_core, curr_patt, new_patt, replaceAll=True)

new_core = Chem.MolToSmiles(Chem.RemoveHs(new_c[0], sanitize=False), kekuleSmiles=True)
```

```python
#Our new core looks like

new_core
```

```python
# Visually

Chem.MolFromSmiles(new_core, sanitize=False)
```

## B. Prepare Rgroup

### &emsp; (i) Attach a dummy atom to the Rgroup 
#### &emsp; &emsp;The dummy atom MUST bear the same pattern as that present in the core

```python

dummy = "[*:1]"

d = Chem.MolFromSmiles(dummy, sanitize=False)

r = Chem.MolFromSmiles(Rgroup, sanitize=False)

#merge molecules
combi = Chem.CombineMols(d,r)

editable = Chem.EditableMol(combi)

editable.AddBond(0,1,Chem.rdchem.BondType.SINGLE) 

new_r = editable.GetMol()

new_Rgroup = Chem.MolToSmiles(Chem.RemoveHs(new_r, sanitize=False), kekuleSmiles=True)
```

```python
# Our Rgroup looks like

new_Rgroup
```

```python
#Visually

Chem.MolFromSmiles(new_Rgroup, sanitize=False)
```

## C. Weld Rgroup onto the core

```python

c = Chem.MolFromSmiles(new_core)     
r = Chem.MolFromSmiles(new_Rgroup)

combined = Chem.CombineMols(c,r)

result = Chem.molzip(combined)

result
```

## Result

```python
# Our resultant molecule looks like

[a.SetAtomMapNum(0) for a in result.GetAtoms()]  #remove atom map numbers

result_SMI = Chem.MolToSmiles(Chem.RemoveHs(result, sanitize=False), kekuleSmiles=True)

print('\n', result_SMI, '\n')

# Visually

result
```

# Method 2


### &emsp; (i) Obtain SMILES of prepared core and Rgroup molecules 

```python
# core looks like

new_core
```

```python
# Rgroup looks like

new_Rgroup
```

#### - A different number/index of the dummy atom in the Rgroup/core may be used. e,g.


```python

#Replacing '[*:1]' with '[*:2]'

curr_patt = Chem.MolFromSmiles('[*:1]') 
new_patt = Chem.MolFromSmiles('[*:2]')   

c = Chem.MolFromSmiles(new_core)     

new_c = Chem.ReplaceSubstructs(c, curr_patt, new_patt, replaceAll=True)

new_core = Chem.MolToSmiles(new_c[0], kekuleSmiles=True)

new_core
```

### &emsp; (ii) Merge core and Rgroup

```python

c = Chem.MolFromSmiles(new_core, sanitize=False)
r = Chem.MolFromSmiles(new_Rgroup, sanitize=False)

combi = Chem.CombineMols(c,r)

combi
```

### &emsp;(iii) Obtain indices of dummy atoms


```python

dc,dr = np.where(np.array([k.GetAtomicNum() for k in combi.GetAtoms()]) == 0)[0]

print(dc)  #dc is for core 
print(dr)  #dr is for Rgroup
```

### &emsp;(iv) Obtain indices of neighboring atoms 


```python

nc = combi.GetAtomWithIdx(int(dc)).GetNeighbors()[0].GetIdx()
nr = combi.GetAtomWithIdx(int(dr)).GetNeighbors()[0].GetIdx()

print(nc)  #nc is for core 
print(nr)  #nr is for Rgroup
```

### &emsp;(v) Remove the dummy atoms 

```python

editable = Chem.EditableMol(combi)
for d in sorted([dc,dr], reverse=True):
    editable.RemoveAtom(int(d))
```

### &emsp;(vi) adjust atom indices of the neighbouring atoms

```python

nc -= int(dc < nc) + int(dr < nc) # Index of the neighbouring atom of the core
nr -= int(dc < nr) + int(dr < nr) # Index of the neighbouring atom of the Rgroup 

print(nc)
print(nr)
```

### &emsp;(vii) Weld core and Rgroup by adding a single bond between neighbouring atoms

```python

editable.AddBond(nc,nr,Chem.rdchem.BondType.SINGLE)

result = editable.GetMol()

result
```

## Result

```python

# Our resultant molecule looks like

[a.SetAtomMapNum(0) for a in result.GetAtoms()]  #remove atom map numbers

result_SMI = Chem.MolToSmiles(Chem.RemoveHs(result, sanitize=False), kekuleSmiles=True)

print('\n', result_SMI, '\n')

# Visually

result
```
